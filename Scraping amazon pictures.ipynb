```python
import os
import selenium
from selenium import webdriver
import time

import io
import requests
from PIL import Image
```

```python
#Install Driver
webdriver.Chrome(executable_path=r"C:\Users\Rahul\Downloads\chromedriver.exe")
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
url = 'https://www.amazon.in/'

driver.get(url)
```

### Sarees

```python
user_input = input('Enter here for the product search: ')

search = driver.find_element_by_id('twotabsearchtextbox')

search.clear()

search.send_keys(user_input)

enter = driver.find_element_by_id('nav-search-submit-button')
enter.click()
```

```python
urls = []


for page in range(0,4):               #for loop for scrapping 3 page

    url = driver.find_elements_by_xpath("//div[@class = 'a-section a-spacing-none s-image-elevated-grid-grey-overlay']/div[1]/span[1]/a[1]/div[1]/img[1]") #scraping urls

    for i in url:

        urls.append(i.get_attribute("src")) #appending the urls of images list

    nxt_button=driver.find_element_by_xpath("//span[@class='s-pagination-strip']//a")#scraping the list of buttons from the page

    nxt_button.click()

    time.sleep(2)
```

```python
urls
```

```python
os.chdir('C:/Users/Rahul/Downloads/google scrpaed/sarees')
baseDir=os.getcwd()

for i, url in enumerate(urls):
    file_name = f"{i:150}.jpg"    
    try:
        image_content = requests.get(url).content

    except Exception as e:
        print(f"ERROR - COULD NOT DOWNLOAD {url} - {e}")

    try:
        image_file = io.BytesIO(image_content)
        image = Image.open(image_file).convert('RGB')
        
        file_path = os.path.join(baseDir, file_name)
        
        with open(file_path, 'wb') as f:
            image.save(f, "JPEG", quality=85)
        print(f"SAVED - {url} - AT: {file_path}")
    except Exception as e:
        print(f"ERROR - COULD NOT SAVE {url} - {e}")
```
